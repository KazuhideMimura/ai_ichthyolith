### 1. 準備

```python
import os
import sys
import time
import math
import random
from glob import glob

import numpy as np
import matplotlib.pyplot as plt

ROOT_DIR = '../'

sys.path.append(ROOT_DIR)
from mrcnn import utils
from mrcnn import visualize
import mrcnn.model as modellib

from functions.utils import load_pickle
from functions.compute_ap import evaluate
%matplotlib inline

MODEL_DIR = '../models'
DETECTIONS_DIR = '../results'
BASE_DIR_NAME = 'logs_dataset_tooth_denticle_with_handpicked_20200728'
MODEL_NAME = 'mask_rcnn_ichthyolith_0080.h5'
TEST_DATASET_DIR = '../data/images/test/'
DEVICE = '/gpu:0'  # /cpu:0 or /gpu:0
TEST_MODE = 'inference'

MODEL_PATH = os.path.join(MODEL_DIR, BASE_DIR_NAME, MODEL_NAME)
DETECTIONS_PATH = os.path.join(DETECTIONS_DIR, BASE_DIR_NAME, MODEL_NAME[-7:-3], 'detections_eval.pkl')
GTS_PATH = os.path.join(DETECTIONS_DIR, 'gts', 'gts.pkl')

print(f'MODEL_PATH      : {MODEL_PATH}')
print(f'DETECTIONS_PATH : {DETECTIONS_PATH}')
print(f'GTS_PATH        : {GTS_PATH}')
```

## 3. 評価
### 3.1 検出数による評価

```python
from functions.process_pickle import extract_by_scores, extract_specific_size

score_threshold = 0
min_size = None
max_size = None
classes = ['bg', 'tooth', 'denticle']

detections = load_pickle(DETECTIONS_PATH)
detections = extract_by_scores(detections, score_threshold)

gts = load_pickle(GTS_PATH)
gts = extract_specific_size(gts, min_size=min_size, max_size=max_size)

num_detections = [0 for _ in range(len(classes))]
num_gts = [0 for _ in range(len(classes))]

for i in range(len(detections)):
    class_ids = detections[i][0]['class_ids']
    gt_class_ids = gts[i][0]['gt_class_id']
    for class_id in class_ids:
        num_detections[class_id] += 1
    for gt_class_id in gt_class_ids:
        num_gts[gt_class_id] += 1

print('detection')
for i, c in enumerate(classes):
    print(f'{c:8}: {num_detections[i]}')

print()
print('gts')
for i, c in enumerate(classes):
    print(f'{c:8}: {num_gts[i]}')
    
```

### 3.2 Recall, Precision, APによる評価

```python
from functions.compute_ap import evaluate
from functions.process_pickle import extract_by_scores, extract_specific_size, convert_detections_labels, convert_gts_labels


# データの読み込み
detections = load_pickle(DETECTIONS_PATH)
gts = load_pickle(GTS_PATH)

# クラスラベルの変換
assign_dict = {1: 1, 2: 2}
num_classes = len(set([v for v in assign_dict.values()]))

if len(assign_dict) != 0:
    detections = convert_detections_labels(detections, assign_dict)
    gts = convert_gts_labels(gts, assign_dict)

class_ids = [i+1 for i in range(num_classes)]

APs = []
precisions = []
recalls = []
mode = 'box'
score_threshold = 0
min_size = None
max_size = None

detections = extract_by_scores(detections, score_threshold)
gts = extract_specific_size(gts, min_size=min_size, max_size=max_size)

for class_id in class_ids:
    AP, precision, recall = evaluate(detections, gts, label=class_id, mode=mode)
    APs.append(AP)
    precisions.append(precision)
    recalls.append(recall)
    print(f'recall    ({mode}) class{class_id:2d}: {recall[-1]}')
    print(f'precision ({mode}) class{class_id:2d}: {precision[-1]}')
    print(f'AP        ({mode}) class{class_id:2d}: {AP}')
    print(recall[-1])
    print(precision[-1])
    print(AP)
APs = np.array(APs)
precisions = np.array(precisions)
recalls = np.array(recalls)
print(f'mAP ({mode}): {np.mean(APs)}')
```

### 3.3.1 PR曲線の描画 (クラス別)

```python
for class_id, AP, precision, recall in zip(class_ids, APs, precisions, recalls):
    print(f'class: {classes[class_id]}')
    m_recall = np.concatenate(([0.], recall, [recall[-1]]))
    m_precision = np.concatenate(([0.], precision, [0.]))
    
    for i in range(m_precision.size - 1, 0, -1):
        m_precision[i-1] = np.maximum(m_precision[i-1], m_precision[i])
    visualize.plot_precision_recall(AP, m_precision, m_recall)
```

### 3.3.2 PR曲線の描画 (重ねて表示)

```python
_, ax = plt.subplots(1)
ax.set_title(f'Precision-Recall Curve. mAP={np.mean(APs):.4}')
ax.set_ylim(0, 1.1)
ax.set_xlim(0, 1.1)

for class_id, AP, precision, recall in zip(class_ids, APs, precisions, recalls):
    m_recall = np.concatenate(([0.], recall, [recall[-1]]))
    m_precision = np.concatenate(([0.], precision, [0.]))
    
    for i in range(m_precision.size - 1, 0, -1):
        m_precision[i-1] = np.maximum(m_precision[i-1], m_precision[i])
    
    ax.plot(m_recall, m_precision, label=classes[class_id])
    ax.legend()

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```

### 3.4 scoreの閾値によるRecall, Precision, APの変化の可視化

```python
from functions.compute_ap import evaluate
from functions.process_pickle import extract_by_scores, extract_specific_size, convert_detections_labels, convert_gts_labels

# データの読み込み
detections = load_pickle(DETECTIONS_PATH)
gts = load_pickle(GTS_PATH)

# クラスラベルの変換
assign_dict = {1: 1, 2: 2}
num_classes = len(set([v for v in assign_dict.values()]))

if len(assign_dict) != 0:
    detections = convert_detections_labels(detections, assign_dict)
    gts = convert_gts_labels(gts, assign_dict)

class_ids = [i+1 for i in range(num_classes)]

APs = []
precisions = []
recalls = []
mode = 'box'
score_thresholds = np.arange(0, 1, 0.05)

for class_id in class_ids:
    sub_APs = []
    sub_precisions = []
    sub_recalls = []
    for score_threshold in score_thresholds:
        sub_detections = extract_by_scores(detections, score_threshold)
        AP, precision, recall = evaluate(sub_detections, gts, label=class_id, mode=mode)
        sub_APs.append(AP)
        sub_precisions.append(precision[-1])
        sub_recalls.append(recall[-1])
    APs.append(sub_APs)
    precisions.append(sub_precisions)
    recalls.append(sub_recalls)

for class_id, AP, precision, recall in zip(class_ids, APs, precisions, recalls):
    _, ax = plt.subplots(1)
    ax.set_title(f'metrics per score_threshold (class: {classes[class_id]})')
    ax.set_ylim(0, 1.1)
    ax.set_xlim(0, 1.1)
    ax.plot(score_thresholds, AP, label='AP')
    ax.plot(score_thresholds, precision, label='precision')
    ax.plot(score_thresholds, recall, label='recall')
    plt.xlabel('score')
    plt.legend()
    
```

## 4. 可視化
### 4.1 検出領域の可視化 (番号によって指定)

```python
import cv2

from functions.process_pickle import get_top_left_coordinate, get_original_slide_name

image_id = 4
slide_dir = '../data/images/test/'
result_path = os.path.join(os.path.dirname(DETECTIONS_PATH), '1179C_24_CC_25_005.pkl')
img_h = 640
img_w = 640
class_names = ['bg', 'tooth', 'denticle']

detections = load_pickle(result_path)
filename = detections[image_id][0]['filename']
bbox = detections[image_id][0]['rois']
masks = detections[image_id][0]['masks']
class_ids = detections[image_id][0]['class_ids']
scores = detections[image_id][0]['scores']

h, w = get_top_left_coordinate(filename)
slide_name = get_original_slide_name(filename)
slide_path = os.path.join(slide_dir, slide_name)
slide_image = cv2.imread(slide_path)
slide_image = cv2.cvtColor(slide_image, cv2.COLOR_BGR2RGB)
image = slide_image[h:h+img_h, w:w+img_w]

visualize.display_instances(image, bbox, masks, class_ids, class_names, scores=scores, title='detection', figsize=[10, 10])
```

### 4.2 検出領域の可視化 (順番に表示)

```python
import cv2
from IPython.display import clear_output

from functions.process_pickle import get_top_left_coordinate, get_original_slide_name, convert_to_rel_rois

slide_dir = '../data/images/test/'
img_h = 640
img_w = 640
class_names = ['bg', 'tooth', 'denticle']

detections = load_pickle(DETECTIONS_PATH)

image_id = 0
pre_filename = ''
slide_h = 0
slide_w = 0

while image_id >= 0 and image_id < len(detections):
    clear_output(wait=True)
    filename = detections[image_id][0]['filename']
    bbox = detections[image_id][0]['rois']
    class_ids = detections[image_id][0]['class_ids']
    scores = detections[image_id][0]['scores']
    masks = np.zeros((img_h, img_w, 1))
    index = 0
    if pre_filename != filename:
        slide_image = cv2.imread(os.path.join(slide_dir, filename))
        slide_image = cv2.cvtColor(slide_image, cv2.COLOR_BGR2RGB)
        slide_h, slide_w, _ = slide_image.shape
        pre_filename = filename
        
    while index >= 0 and index < len(class_ids):
        clear_output(wait=True)
        image_center_x = int((bbox[index][1] + bbox[index][3]) / 2) 
        image_center_y = int((bbox[index][0] + bbox[index][2]) / 2)
        top_x = int(min(slide_w - img_w, max(0, image_center_x - img_h / 2)))
        top_y = int(min(slide_h - img_h, max(0, image_center_y - img_w / 2)))
        image = slide_image[top_y:top_y+img_h, top_x:top_x+img_w]
        rois = convert_to_rel_rois(np.array([bbox[index]]), top_x, top_y)
        
        visualize.display_instances(image, rois, masks, np.array([class_ids[index]]), 
                                    class_names, scores=np.array([scores[index]]), title='detections', figsize=[6, 6])
        print(f'b … 前に戻る / n … 次に進む / q … 終了 ({index+1:4}/{len(class_ids):4}) {filename}')
        print()
        message = input()
        if message == 'b':
            if index == 0:
                index = -1
                image_id -= 1
                break
            else:
                index -= 1
        elif message == 'n':
            if index == len(class_ids) - 1:
                index = -1
                image_id += 1
                break
            else:
                index += 1
        elif message == 'q':
            image_id = -1
            break
        else:
            continue

print()
print('表示終了')
```

### 4.3 bboxサイズ分布ヒストグラムの可視化

```python
detections = load_pickle(DETECTIONS_PATH)
gts = load_pickle(GTS_PATH)

size_regulation = 640
max_size = 0.05
bbox_size = []
gt_bbox_size = []
for i, detection in enumerate(detections):
    for r in detection[0]['rois']:
        size = ((r[3] - r[1]) / size_regulation) * ((r[2] - r[0]) / size_regulation)
        bbox_size.append(size)


for i, gt in enumerate(gts):
    for g in gt[0]['gt_bbox']:
        size = ((g[3] - g[1]) / size_regulation) * ((g[2] - g[0]) / size_regulation)
        gt_bbox_size.append(size)


fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
ax1.set_title('detections')
ax2.set_title('gts')
ax1.set_xlim((0, max_size))
ax2.set_xlim((0, max_size))
ax1.hist(np.array(bbox_size), rwidth=0.8, bins=100)
ax2.hist(np.array(gt_bbox_size), rwidth=0.8, bins=100)
```
